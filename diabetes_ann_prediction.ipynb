```python
from sklearn.model_selection import train_test_split
```

```python
X= df.drop('Class variable', axis=1)
```

```python
y=df['Class variable']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.25, random_state=42, stratify=y)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train  =scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras.optimizers import RMSprop
```

```python
model = Sequential()
```

```python
model.add(Dense(12, activation = 'relu', input_dim = X_train.shape[1]))
```

```python
model.add(Dense(1,activation = 'sigmoid'))
```

```python
model_optimizer = RMSprop(learning_rate = 0.001)
model.compile(optimizer = model_optimizer, loss ='binary_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

```python
train_model = model.fit( X_train, y_train, epochs=20, batch_size =128, validation_data =(X_test,y_test))
```

```python
plt.plot(train_model.history['loss'], label='Train Loss')
plt.plot(train_model.history['val_loss'], label='Test Loss')
plt.legend(); plt.title('Loss Curve');
plt.show()
```

```python
plt.plot(train_model.history['accuracy'], label='Train accuracy')
plt.plot(train_model.history['val_accuracy'], label='Test accuracy')
plt.legend(); plt.title('accuracy Curve');
plt.show()
```

```python
train_loss, train_accuracy= model.evaluate(X_train, y_train, verbose= 0)
```

```python
test_loss, test_accuracy= model.evaluate(X_test, y_test, verbose= 0)
```

```python
#Adams model
from tensorflow.keras.optimizers import Adam
model_1 = Sequential([Input(shape =(X_train.shape[1],)), Dense(12, activation ='relu'),Dense(1, activation='sigmoid')])
```

```python
adam_optimizer = Adam(learning_rate = 0.001)
```

```python
model_1.compile(optimizer = adam_optimizer , loss ='binary_crossentropy', metrics=['accuracy'])
```

```python
model_1_result = model_1.fit(X_train, y_train , epochs =40, batch_size = 128, validation_data=(X_test,y_test), verbose= 1)
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
```

```python
adam_roc = model_1.predict(X_test).ravel() #In keras we use ravel()
```

```python
fpr, tpr, _ = roc_curve(y_test,adam_roc)
```

```python
adam_roc_auc = roc_auc_score(y_test, adam_roc)
```

```python
plt.figure()
plt.plot(fpr,tpr)
plt.plot([0,1],[0,1])
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('Adam ROC curve')
```

```python
adam_auc = roc_auc_score(y_test,adam_roc)
print(adam_auc)
```

```python
# Model 2 SDG(stocchastic gradient descent)
from tensorflow.keras.optimizers import SGD
model_2 = Sequential([Input(shape =(X_train.shape[1],)), Dense(12, activation ='relu'),Dense(1, activation='sigmoid')])
model_2.compile(optimizer = SGD(learning_rate = 0.01),loss='binary_crossentropy', metrics=['accuracy'])
```

```python
sgd_moedl_2 = model_2.fit(X_train,y_train, epochs =20, batch_size = 10, validation_data=(X_test,y_test), verbose= 1)
# I tried batch_size with 128,the accuracy only hits 69%. After tuning the size to 10, about 10 % increase.
```

```python
SGD_roc = model_2.predict(X_test).ravel()
```

```python
sgd_fpr, sgd_tpr, _ = roc_curve(y_test,SGD_roc)
SGD_roc_auc = roc_auc_score(y_test, SGD_roc)
```

```python
plt.figure()
plt.plot(sgd_fpr,sgd_tpr)
plt.plot([0,1],[0,1])
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('SGD ROC curve')
```

```python
from tensorflow.keras.optimizers import Ftrl
model_4 = Sequential([Input(shape =(X_train.shape[1],)), Dense(12, activation ='relu'),Dense(1, activation='sigmoid')])
model_4.compile(optimizer = Ftrl(learning_rate = 0.01),loss='binary_crossentropy', metrics=['accuracy'])
```

```python
Ftrl_auc = roc_auc_score(y_test,Ftrl_roc)
print(Ftrl_auc)
```
